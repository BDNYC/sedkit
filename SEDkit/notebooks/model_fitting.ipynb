```python
# Imports
from SEDkit import Vega, SED, Spectrum, Blackbody, VegaSED, BTSettl, FileSpectrum
import astropy.units as q
import numpy as np
from pkg_resources import resource_filename
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
output_notebook()
```

# Fitting models to a `Spectrum`

```python
# Make the SED
s = SED(name='2MASSW J0036159+182110')
s.add_photometry_file(resource_filename('SEDkit', 'data/L3_photometry.txt'))
s.find_WISE()
s.find_Gaia()
s.spectral_type = 'L3.5'
file = resource_filename('SEDkit', 'data/L3_spectrum.fits')
s.add_spectrum_file(file, wave_units=q.um, flux_units=q.W/q.m**2/q.um)
```

```python
# s.fit_blackbody(fit_to='app_phot_SED')
s.plot(integral=True)
```

```python
vega = VegaSED()
vega.plot(integral=True)
```
