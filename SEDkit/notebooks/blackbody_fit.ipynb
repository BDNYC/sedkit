```python
# Imports
from SEDkit import sed, spectrum, synphot, utilities
import astropy.units as q
import numpy as np
from pkg_resources import resource_filename
from bokeh.io import output_notebook, show
output_notebook()
```

```python
# Make the SED
s = sed.SED(name='2MASSW J0036159+182110')
s.add_photometry_file(resource_filename('SEDkit', 'data/L3_photometry.txt'))
s.find_WISE()
s.find_Gaia()
s.spectral_type = 'L3.5'
```

```python
# Add a blackbody segment as the spectrum
bb_wave = np.linspace(1., 2.5, 200)*q.um
bb_flux, bb_unc = utilities.blackbody(bb_wave, 1817*q.K, 100*q.K)
s.add_spectrum([bb_wave, bb_flux, bb_unc])
```

```python
s.fit_blackbody(fit_to='app_SED')
s.plot()
```
