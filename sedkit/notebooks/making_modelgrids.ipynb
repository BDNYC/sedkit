# Make your own `ModelGrid` object
Here you can turn a bunch of arbitrary numpy arrays into a `ModelGrid` object.

First, create an empty instance with the desired parameters.

```python
mg = ModelGrid('Filippazzo+2016', ['spectral_type', 'teff'], wave_units=q.um, flux_units=q.erg/q.s/q.cm**2/q.AA)
```

To save all this work we are doing, let's make a directory of XML files from our spectra that can be loaded in very fast later.

```python
# Grab all SEDs
files = glob('/Users/jfilippazzo/Documents/CUNY/Dissertation/Sample/absolute SEDs/*.txt')
for fname in files:
    with open(fname) as f:
        
        # Read the header
        content = [l.split('=') for l in f.readlines() if l.startswith('#') and '=' in l]
        d = {k[0].split(' ')[1]:k[1].split(' ')[0].replace('\n','') for k in content}
        
    # Get the spectrum
    spec = np.genfromtxt(fname, unpack=True)
    
    # Add the object
    mg.add_model(spec, **d)
```
