# Make a catalog of SEDs from a file of coordinates
You can also create an SED Catalog by passing a file of `ra` and `dec` values (in decimal degrees) to the `SED.from_coord_list()` method like so:

```python
# Imports
from SEDkit import sed, spectrum, synphot, catalog
import astropy.units as q
import numpy as np
from pkg_resources import resource_filename
from bokeh.io import output_notebook, show
output_notebook()
```

```python
# Make a new catalog
cat = catalog.SEDCatalog()
```

```python
# Read in the data
cat.from_file(resource_filename('SEDkit', 'data/coords.txt'))
```

```python
# Check out the results
cat.results
```

```python
# Plot some stuff
show(cat.plot('2MASS.J-2MASS.Ks', 'Teff'))
```
